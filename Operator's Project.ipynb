# Arithmatic Operators

***Add (+)***

```python
# Add (+) Values.
x = 10 #int
y = 5 #int
print(x + y)
```

***Subtract (-)***

```python
# Subtract (-) Values.
x = 15 #int
y = 5 #int
print(x - y)
```

***Multiply ()***

```python
# Multiply (*) values.
x = 6 #int
y = 3 #int
print(x * y)
```

***Divide ( / )***

```python
# Divide (/) give a float values.
x = 30 #int
y = 3 #int
print(x / y)
```

***Floor Division ( // )***

```python
# Floor Divide (//) give a integer Values.
x = 27 #int
y = 5 #int
print(x // y)
```

***Moduls Division ( % )***

```python
# Modulas Division ( % ) give a reminder values.
x = 50 #int
y = 6 #int
print(x % y)
```

***Exponentiation ()***

```python
# Exponentiation (**) values.
x = 6**2 #int
print(x)
```

## Comparison Operators

***Equal to (==)***

```python
# Equal to (==) 
#check both values is Equal.
x = 8 #int
y = 8 #int
print(x==y)
```

***Not Equal to (!=)***

```python
# Not Equal (!=) 
# check both values is not Equal.
x = 7 #int
y = 7 #int
print(x!=y)
```

***Greater than (>)***

## Logical Operators

***And Operator***

```python
# And operators.
x = 10
y = 5
# Both conditions are true than print true.
print(x>5 and y<10) 
```

```python
# And operators.
x = 10
y = 5
# Both conditions are not true than print false.
print(x>15 and y<10) 
```

```python
# And operators.
x = 10
y = 5
# Both conditions are not true than print false.
print(x>15 and y<17) 
```

***Or Operator***

```python
# Or operators.
x = 15
y = 10
# If a condition is true than print True.
print(x>10 or y<9)
```

```python
# Or operators.
x = 15
y = 10
# Both conditions are true than print true.
print(x>10 or y<11)
```

```python
# Or operators.
x = 15
y = 10
# Both conditions are false than print False.
print(x>20 or y<5)
```
