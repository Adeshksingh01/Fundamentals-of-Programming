```python
#1.Check if a number is even and odd.
num = (int(input("Enter a number:")))
if num % 2 == 0:
    print(f'{num}: is a even number.')
else:
    print(f'{num}: is a odd number.')
```

```python
#2.Check if a person is eligible to vote (age >= 18).
age = (int(input("Enter a number:")))
if age >= 18:
    print("A person is eligible to vote")
else:
    print("A person is not eligiable to vote")
```

```python
#3.Find the greatest of two numbers.
a = (int(input("Enter a first number:")))
b = (int(input("Enter a second number:")))
if a > b :
    print(f'{a}: is greater than :{b}')
else:
    print(f'{b}: is greater than :{a}')
```

```python
#4.Check if a number is divisible by 5.
num = (int(input("Enter a number:")))
if num % 5 == 0:
    print(f'{num}: is divisible by 5')
else:
    print(f'{num}: is not divisible by 5')
```

```python
#5.Check if a number is within a range (20 to 60).
num = (int(input("Enter a number:")))
if num > 20 and num < 60:
    print(f'{num}: is within a range (20 to 60)')
else:
    print(f'{num}: is not within a range (20 to 60)')
```

```python
#6.Check if a number is multiple of both 5 and 9.
num = (int(input("Enter a number:")))
if num % 5 == 0 and num % 9 == 0:
    print(f'{num}: is multiple of both 5 and 9.')
else:
    print(f'{num}: is not multiple of both 5 and 9.') 
```

```python
#7.Check if a day is weekday or weekend
weekday = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
day = input("Enter a day: ")
if day in weekday:
    print(f"{day} is a weekday.")
else:
    print(f"{day} is a weekend.")
```

```python
#8.Find profit or loss.
x = int(input("Enter a MRP:"))
y = int(input("Enter a selling price:"))
if x < y:
    print(f"Profit of {y-x}.")
else:
    print(f"Loss of {x-y}.")
```
