```python
# Create a function for calculates the square of a number.
def square(num):
    return num * num

n = int(input("Enter a number: "))
print("Square:", square(n))
```

```python
# Create a function for calculates the factorial of 1 to n numbers.
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

num = int(input("Enter a number: "))
print("Factorial:", factorial(num))
```

```python
# Create a function for check numbers is Even and Odd.
def check_even_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

n = int(input("Enter a number: "))
print("The number is", check_even_odd(n))
```

```python
# Create a function for find the Sum and Average of numbers.
def calculate(a, b):
    total = a + b
    avg = total / 2
    return total, avg

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
s, a = calculate(x, y)
print("Sum =", s, "Average =", a)
```

```python
# Create a function for count the numbers of vowels in the sentence.
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

s = input("Enter a string: ")
print("Number of vowels:", count_vowels(s))
```

```python
# Create a function for find the simple intrest.
def simple_interest(p, r, t):
    return (p * r * t) / 100
```

```python
print(simple_interest(1000, 5, 2))
```

```python
print(simple_interest(200, 6, 8))
```

```python
# Create a function for get a even number.
def get_even_numbers(lst):
    return [x for x in lst if x % 2 == 0]
```

```python
print(get_even_numbers([1, 2, 3, 4, 5, 6]))
```

```python
# Create a funcion for count how many word in the sentence.
def count_words(text):
    return len(text.split())
print(count_words("Python is easy to learn"))
```

```python
# Create a function for conver C into F.
def c_to_f(c):
    return (c * 9/5) + 32
print(c_to_f(25))
```

```python
# Create a function for find the key name which value is max.
def max_value_key(d):
    return max(d, key=d.get)
print(max_value_key({'a': 10, 'b': 20, 'c': 5}))
```

```python
# Create a function for count the words of the paragraphs.
def word_count(paragraph, word):
    return paragraph.lower().split().count(word.lower())
print(word_count("Python is fun. I love Python.", "python"))
```

```python
# Create a function for find the average of the list.
def average_marks(marks):
    return sum(marks) / len(marks)
print(average_marks([90, 80, 70, 85]))
```

```python
# Create a function for find the square of the list.
def square_list(lst):
    return [x * x for x in lst]
print(square_list([1, 2, 3, 4]))
```

```python
# Create a function for check login successful or unsuccessful.
def login(user, password):
    if user == "admin" and password == "1234":
        return "Login Successful"
    else:
        return "Invalid Credentials"
print(login("admin", "1234"))
```

```python
# Create a function for check a years is a leap year or not.
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return "Leap Year"
    else:
        return "Not a Leap Year"
print(is_leap_year(2024))
```
