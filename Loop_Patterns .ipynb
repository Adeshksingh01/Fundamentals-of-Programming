```python
# Right-Angled Triangle Star Pattern (Increasing Order).
""" This code prints stars from 1 to 10 in increasing order -- forming traingle that grows downwards. """
for i in range(0,11):
    for j in range(1,i+1):
        print("*",end=" ")
    print()
```

```python
# Right-Angled Triangle Star Pattern (Decreasing Order).
""" This code prints star from 10 to 1 in decreasing order -- forming an inverted triangle. """
for i in range(10,0,-1):
    for j in range(i):
        print("*",end=" ")
    print()
```

```python
# Write a python program to print an Upright Pyramid Pattern.
n = int(input("Enter a range:"))
for i in range(1,n+1):
    print(" "* (n-i) + "*" * (2*i-1))
```

```python
# Write a python program to print an Inverted Pyramid Pattern.
n = int(input("Enter a range:"))
for i in range(n,0,-1):
    print(" " * (n-i) + "*" * (2*i-1))
```

```python
# Write a python program to generate a diamond pattern using for loop.
n = int(input("Enter a range:"))
#Upper part of diamond
for i in range(1,n+1):
    print(" " * (n-i) + "*" * (2*i-1))
#Lower part of diamond
for i in range(n-1,0,-1):
    print(" " * (n-i) + "*" * (2*i-1))
print()
```

```python
# Write a python program to generate a diamond pattern using for loop.
rows = int(input("Enter a range:"))
# Upper part
for i in range(1,rows+1):
    print(" " * (rows-i) + "*" + " " * (2 * i-3) + ("*" if i > 1 else " " ))
# Lower part
for i in range(rows-1,0,-1):
    print(" " * (rows-i) + "*" + " " * (2 * i-3) + ("*" if i > 1 else " " ))
```

```python
# Write a python program to create pattern using nasted loops.
rows = int(input("Enter a range:"))
# Upper part
for i in range(1,rows + 1):
    print("*" * i + " " * (2 * (rows - i)) + "*" * i)
# Lower part
for i in range(rows-1,0,-1):
    print("*" * i + " " * (2 * (rows - i)) + "*" * i)
print()
```

```python
# Write a python program to create pattern using nasted loops.
rows = int(input("Enter a range:"))
for i in range(1,rows+1):
    for j in range(1,rows+1):
        if i == j or i + j == rows + 1:
            print("*",end="")
        else:
            print(" ",end="")
    print()
```

```python
# Write a python program to create pattern using nasted loops.
n = int(input("Enter a range:"))
for i in range(n):
    p = 1
    for j in range(i,n):
        print(" ",end=" ")
    for j in range(i):
        print(p,end=" ")
        p+=1
    for j in range(i+1):
        print(p,end=" ")
        p-=1
    print()
```

```python
# Write a python program to generate a square pattern using for loop.
def hollow_square(n):
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n-1 or j == 0 or j == n-1:
                print("*",end=" ")
            else:
                print(" ",end=" ")
        print()
```

```python
hollow_square(10)
```

```python
# Write a python program to generate a Bordered Name pattern.
def my_name(n):
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n-1 or j == 0 or j == n-1:
                print("*",end=" ")
            elif i == 3 and j == 1:
                print("A",end=" ")
            elif i == 3 and j == 2:
                print("d",end=" ")
            elif i == 3 and j == 3:
                print("e",end=" ")
            elif i == 3 and j == 4:
                print("s",end=" ")
            elif i == 3 and j == 5:
                print("h",end=" ")
            else:
                print(" ",end=" ")
        print()
```

```python
my_name(7)
```
